```python
import re

def add_quotes_to_json(s):
    # This regular expression finds key: value pairs where the key isn't quoted
    pattern = r'(?<!")(\b[a-zA-Z_]\w*\b)(?!":)(?=:)'
    # Replace unquoted keys with quoted keys
    quoted = re.sub(pattern, r'"\1"', s)
    return quoted

# Test
fixed_json = add_quotes_to_json(invalid_json)
print(fixed_json)  # Should print: {"name": "John", "age": 30, "city": "New York"}
```
